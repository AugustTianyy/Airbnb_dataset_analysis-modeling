```python
def filter_outliers(data, column, multiplier=1.5):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    
    print(f"{column}: Lower bound = {lower_bound}, Upper bound = {upper_bound}")
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
```

```python
# filter multiple variables
columns_to_filter = ['price', 'number_of_reviews', 'availability_365']
for col in columns_to_filter:
    data = filter_outliers(data, col)
```

```python
import numpy as np
from scipy.stats import expon, gamma, lognorm
import matplotlib.pyplot as plt

# Data
number_of_reviews = data['number_of_reviews']

# Fit an exponential distribution
params_exp = expon.fit(number_of_reviews)  # Fit parameters
x = np.linspace(number_of_reviews.min(), number_of_reviews.max(), 100)
pdf_exp = expon.pdf(x, *params_exp)

# Fit a gamma distribution
params_gamma = gamma.fit(number_of_reviews)
pdf_gamma = gamma.pdf(x, *params_gamma)

# Fit a log-normal distribution
params_lognorm = lognorm.fit(number_of_reviews)
pdf_lognorm = lognorm.pdf(x, *params_lognorm)

# Plot
plt.hist(number_of_reviews, bins=30, density=True, alpha=0.5, label='Data')

plt.plot(x, pdf_exp, label='Exponential', color='blue')
plt.plot(x, pdf_gamma, label='Gamma', color='green')
plt.plot(x, pdf_lognorm, label='Log-Normal', color='red')

plt.legend()
plt.title('Distribution Fitting')
plt.ylim(0, 0.6)  # Limit the Y-axis range to 0-0.6

plt.show()
```

```python
# show only the hist diagram
plt.hist(number_of_reviews, bins=30, density=True, alpha=0.5, label='Data')
plt.show()
```

```python
print("Exponential Fit Params:", params_exp)
print("Gamma Fit Params:", params_gamma)
print("Log-Normal Fit Params:", params_lognorm)
```

```python
from scipy.stats import gamma, expon, lognorm

ll_gamma = gamma.logpdf(number_of_reviews, *params_gamma).sum()
ll_expon = expon.logpdf(number_of_reviews, *params_exp).sum()
ll_lognorm = lognorm.logpdf(number_of_reviews, *params_lognorm).sum()

print("Log-Likelihood:")
print(f"Gamma: {ll_gamma}, Exponential: {ll_expon}, Log-Normal: {ll_lognorm}")

# the result indicates that log-normal is the best fit for now, bruhhhhhhh.....
```

```python
n_params_gamma = len(params_gamma)  # Gamma distribution parameters
n_params_expon = len(params_exp)   # Exponential distribution parameters
n_params_lognorm = len(params_lognorm)  # Log-normal distribution parameters

# Number of data points
n = len(number_of_reviews)

# AIC Calculation
aic_gamma = 2 * n_params_gamma - 2 * ll_gamma
aic_expon = 2 * n_params_expon - 2 * ll_expon
aic_lognorm = 2 * n_params_lognorm - 2 * ll_lognorm

# BIC Calculation
bic_gamma = n_params_gamma * np.log(n) - 2 * ll_gamma
bic_expon = n_params_expon * np.log(n) - 2 * ll_expon
bic_lognorm = n_params_lognorm * np.log(n) - 2 * ll_lognorm

print("AIC:")
print(f"Gamma: {aic_gamma}, Exponential: {aic_expon}, Log-Normal: {aic_lognorm}")

print("BIC:")
print(f"Gamma: {bic_gamma}, Exponential: {bic_expon}, Log-Normal: {bic_lognorm}")


# the lower the better (more negative is better), which indicates lognormal also better under this standard, bruhhhhhh......
```

```python
from scipy.stats import kstest

ks_gamma = kstest(number_of_reviews, 'gamma', args=params_gamma)
ks_expon = kstest(number_of_reviews, 'expon', args=params_exp)
ks_lognorm = kstest(number_of_reviews, 'lognorm', args=params_lognorm)

print("K-S Test:")
print(f"Gamma: {ks_gamma}, Exponential: {ks_expon}, Log-Normal: {ks_lognorm}")

# okay, now it indicates that none of those fit
```

```python
import matplotlib.pyplot as plt
from scipy.stats import gamma, expon, lognorm

# ECDF
sorted_data = np.sort(number_of_reviews)
ecdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Theoretical CDFs
x = np.linspace(sorted_data.min(), sorted_data.max(), 500)
cdf_gamma = gamma.cdf(x, *params_gamma)
cdf_expon = expon.cdf(x, *params_exp)
cdf_lognorm = lognorm.cdf(x, *params_lognorm)

# Plot
plt.figure(figsize=(10, 6))
plt.step(sorted_data, ecdf, label='Empirical CDF', color='black')
plt.plot(x, cdf_gamma, label='Gamma CDF', color='blue')
plt.plot(x, cdf_expon, label='Exponential CDF', color='green')
plt.plot(x, cdf_lognorm, label='Log-Normal CDF', color='red')
plt.legend()
plt.title('Empirical vs. Theoretical CDFs')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from keras_tuner.tuners import RandomSearch
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


# Normalize the target variable y
y_scaler = MinMaxScaler()

y_train = pd.Series(y_train)
y_test = pd.Series(y_test)
y_train_scaled = y_scaler.fit_transform(y_train.to_numpy().reshape(-1, 1))  # Reshape y into 2D array for normalization
y_test_scaled = y_scaler.transform(y_test.to_numpy().reshape(-1, 1))

# Define model building function
def build_model(hp):
    model = Sequential([
        Dense(
            units=hp.Int('units_1', min_value=32, max_value=128, step=32),
            activation='relu',
            input_dim=X_train.shape[1]
        ),
        Dropout(hp.Float('dropout_rate', min_value=0.1, max_value=0.5, step=0.1)),
        Dense(
            units=hp.Int('units_2', min_value=16, max_value=64, step=16),
            activation='relu'
        ),
        Dense(1)  # Output layer
    ])
    model.compile(
        optimizer=hp.Choice('optimizer', values=['adam', 'sgd']),
        loss='mse',
        metrics=['mse']
    )
    return model

# Initialize Keras Tuner
new_tuner = RandomSearch(
    build_model,                         # Model building function
    objective='mse',                     # Optimization target
    max_trials=10,                       # Maximum number of combinations to search
    executions_per_trial=3,              # Number of repetitions per hyperparameter combination
    directory='my_new1_tuner',           # Directory to store results
    project_name='new1_hyperparam_search', # Project name
    max_consecutive_failed_trials=5      # Raise threshold
)

class CustomVerboseCallback(tf.keras.callbacks.Callback):
    def __init__(self, scaler, X_train, y_train_scaled):
        super().__init__()
        self.scaler = scaler
        self.scale_factor = (scaler.data_max_[0] - scaler.data_min_[0]) ** 2
        self.X_train = X_train
        self.y_train_scaled = y_train_scaled

    def on_epoch_end(self, epoch, logs=None):
        loss_scaled = logs.get('loss')
        val_loss_scaled = logs.get('val_loss')
        mse_scaled = logs.get('mse')
        val_mse_scaled = logs.get('val_mse')

        # Restore loss values
        if loss_scaled is not None and val_loss_scaled is not None:
            loss_original = loss_scaled * self.scale_factor
            val_loss_original = val_loss_scaled * self.scale_factor
            print(f"Epoch {epoch + 1}: Loss (Original Scale): {loss_original:.4f}, Validation Loss (Original Scale): {val_loss_original:.4f}", end="")

        # Restore MSE
        if mse_scaled is not None and val_mse_scaled is not None:
            mse_original = mse_scaled * self.scale_factor
            val_mse_original = val_mse_scaled * self.scale_factor
            print(f", MSE (Original Scale): {mse_original:.4f}, Validation MSE (Original Scale): {val_mse_original:.4f}", end="")

        # Train RMSE and R²
        if self.model is not None:
            y_train_pred_scaled = self.model.predict(self.X_train, verbose=0)
            y_train_pred = self.scaler.inverse_transform(y_train_pred_scaled)
            y_train_true = self.scaler.inverse_transform(self.y_train_scaled)

            train_rmse = np.sqrt(np.mean((y_train_true - y_train_pred) ** 2))
            train_r2 = 1 - np.sum((y_train_true - y_train_pred) ** 2) / np.sum((y_train_true - np.mean(y_train_true)) ** 2)
            print(f", Train RMSE: {train_rmse:.4f}, Train R²: {train_r2:.4f}")

# Create callback instance
custom_verbose_callback = CustomVerboseCallback(
    y_scaler, 
    X_train, 
    y_train_scaled
)

# Start hyperparameter tuning
new_tuner.search(
    X_train,
    y_train_scaled,
    epochs=50,
    validation_split=0.2,  # Automatically split validation set
    verbose=0,  # Disable default output
    callbacks=[custom_verbose_callback]  # Add custom callback
)

# Get the best hyperparameters
# This line won't be fixed by changing the tuner name, only the following line is correct
best_hps = new_tuner.get_best_hyperparameters(num_trials=1)[0]

# Output the best hyperparameters
print("Best Hyperparameters:")
for key, value in best_hps.values.items():
    print(f"{key}: {value}")

# Train the final model using the best hyperparameters
best_model = new_tuner.hypermodel.build(best_hps)

# Define ScaleLossCallback if needed
class ScaleLossCallback(tf.keras.callbacks.Callback):
    def __init__(self, scaler):
        super().__init__()
        self.scaler = scaler
        self.scale_factor = (scaler.data_max_[0] - scaler.data_min_[0]) ** 2

    def on_epoch_end(self, epoch, logs=None):
        loss_scaled = logs.get('loss')
        val_loss_scaled = logs.get('val_loss')
        mse_scaled = logs.get('mse')
        val_mse_scaled = logs.get('val_mse')

        if loss_scaled is not None and val_loss_scaled is not None:
            loss_original = loss_scaled * self.scale_factor
            val_loss_original = val_loss_scaled * self.scale_factor
            print(f"Epoch {epoch + 1}: Loss (Original Scale): {loss_original:.4f}, Validation Loss (Original Scale): {val_loss_original:.4f}", end="")
        if mse_scaled is not None and val_mse_scaled is not None:
            mse_original = mse_scaled * self.scale_factor
            val_mse_original = val_mse_scaled * self.scale_factor
            print(f", MSE (Original Scale): {mse_original:.4f}, Validation MSE (Original Scale): {val_mse_original:.4f}")

# Train the model
scale_loss_callback = ScaleLossCallback(y_scaler)
history = best_model.fit(
    X_train, 
    y_train_scaled, 
    epochs=50, 
    batch_size=32, 
    validation_split=0.2, 
    verbose=0, 
    callbacks=[scale_loss_callback]  # Add callback
)

# Test set evaluation
y_pred_scaled = best_model.predict(X_test)

# Restore predicted and target values
y_pred = y_scaler.inverse_transform(y_pred_scaled)

# Compute MSE and R²
mse = np.mean((y_test - y_pred.squeeze())**2)
r2 = 1 - np.sum((y_test - y_pred.squeeze())**2) / np.sum((y_test - np.mean(y_test))**2)

print(f"Test MSE: {mse:.4f}")
print(f"Test R²: {r2:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from keras_tuner.tuners import RandomSearch
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# === Data preparation starts ===
np.random.seed(42)

X_linear = np.random.rand(1000, 3)
noise_linear = np.random.normal(0, 0.1, size=(1000,))
y_linear = 3 * X_linear[:, 0] + 2 * X_linear[:, 1] - X_linear[:, 2] + noise_linear
y_linear = y_linear.reshape(-1, 1)

X_train_linear, X_test_linear, y_train_linear, y_test_linear = train_test_split(X_linear, y_linear, test_size=0.2, random_state=42)

y_scaler_linear = MinMaxScaler()
y_train_scaled_linear = y_scaler_linear.fit_transform(y_train_linear)  
y_test_scaled_linear = y_scaler_linear.transform(y_test_linear)
# === Data preparation ends ===

def build_model_linear(hp):
    try:
        model = Sequential([
            Dense(
                units=hp.Int('units_1', min_value=32, max_value=128, step=32),
                activation='relu',
                input_dim=X_train_linear.shape[1]
            ),
            Dropout(hp.Float('dropout_rate', min_value=0.1, max_value=0.5, step=0.1)),
            Dense(
                units=hp.Int('units_2', min_value=16, max_value=64, step=16),
                activation='relu'
            ),
            Dense(1)
        ])
        model.compile(
            optimizer=hp.Choice('optimizer', values=['adam', 'sgd']),
            loss='mse',
            metrics=['mse']
        )
        return model
    except Exception as e:
        print(f"Error in build_model_linear: {e}")
        raise

# Initialize Keras Tuner
tuner_linear = RandomSearch(
    build_model_linear,
    objective='mse',
    max_trials=10,
    executions_per_trial=3,
    directory='my_tuner_linear',
    project_name='hyperparam_search_linear',
    max_consecutive_failed_trials=5  # Increase threshold
)

class CustomVerboseCallbackLinear(tf.keras.callbacks.Callback):
    def __init__(self, scaler, X_train, y_train_scaled):
        super().__init__()
        self.scaler = scaler
        self.scale_factor = (scaler.data_max_[0] - scaler.data_min_[0]) ** 2
        self.X_train = X_train
        self.y_train_scaled = y_train_scaled

    def on_epoch_end(self, epoch, logs=None):
        loss_scaled = logs.get('loss')
        val_loss_scaled = logs.get('val_loss')
        mse_scaled = logs.get('mse')
        val_mse_scaled = logs.get('val_mse')

        # Restore loss values
        if loss_scaled is not None and val_loss_scaled is not None:
            loss_original = loss_scaled * self.scale_factor
            val_loss_original = val_loss_scaled * self.scale_factor
            print(f"Epoch {epoch + 1}: Loss (Original Scale): {loss_original:.4f}, Validation Loss (Original Scale): {val_loss_original:.4f}", end="")

        # Restore MSE
        if mse_scaled is not None and val_mse_scaled is not None:
            mse_original = mse_scaled * self.scale_factor
            val_mse_original = val_mse_scaled * self.scale_factor
            print(f", MSE (Original Scale): {mse_original:.4f}, Validation MSE (Original Scale): {val_mse_original:.4f}", end="")

        # Train RMSE and R²
        if self.model is not None:
            y_train_pred_scaled = self.model.predict(self.X_train, verbose=0)
            y_train_pred = self.scaler.inverse_transform(y_train_pred_scaled)
            y_train_true = self.scaler.inverse_transform(self.y_train_scaled)

            train_rmse = np.sqrt(np.mean((y_train_true - y_train_pred) ** 2))
            train_r2 = 1 - np.sum((y_train_true - y_train_pred) ** 2) / np.sum((y_train_true - np.mean(y_train_true)) ** 2)
            print(f", Train RMSE: {train_rmse:.4f}, Train R²: {train_r2:.4f}")

# Create callback instance
custom_verbose_callback_linear = CustomVerboseCallbackLinear(
    y_scaler_linear, 
    X_train_linear, 
    y_train_scaled_linear
)

# Start hyperparameter tuning
tuner_linear.search(
    X_train_linear,
    y_train_scaled_linear,
    epochs=50,
    validation_split=0.2,  
    verbose=0,  
    callbacks=[custom_verbose_callback_linear]  
)

# Get the best hyperparameters
best_hps_linear = tuner_linear.get_best_hyperparameters(num_trials=1)[0]

print("Best Hyperparameters:")
for key, value in best_hps_linear.values.items():
    print(f"{key}: {value}")

# Train the final model using the best hyperparameters
best_model_linear = tuner_linear.hypermodel.build(best_hps_linear)

# Define ScaleLossCallback if needed
class ScaleLossCallbackLinear(tf.keras.callbacks.Callback):
    def __init__(self, scaler):
        super().__init__()
        self.scaler = scaler
        self.scale_factor = (scaler.data_max_[0] - scaler.data_min_[0]) ** 2

    def on_epoch_end(self, epoch, logs=None):
        loss_scaled = logs.get('loss')
        val_loss_scaled = logs.get('val_loss')
        mse_scaled = logs.get('mse')
        val_mse_scaled = logs.get('val_mse')

        if loss_scaled is not None and val_loss_scaled is not None:
            loss_original = loss_scaled * self.scale_factor
            val_loss_original = val_loss_scaled * self.scale_factor
            print(f"Epoch {epoch + 1}: Loss (Original Scale): {loss_original:.4f}, Validation Loss (Original Scale): {val_loss_original:.4f}", end="")
        if mse_scaled is not None and val_mse_scaled is not None:
            mse_original = mse_scaled * self.scale_factor
            val_mse_original = val_mse_scaled * self.scale_factor
            print(f", MSE (Original Scale): {mse_original:.4f}, Validation MSE (Original Scale): {val_mse_original:.4f}")

scale_loss_callback_linear = ScaleLossCallbackLinear(y_scaler_linear)
history_linear = best_model_linear.fit(
    X_train_linear, 
    y_train_scaled_linear, 
    epochs=50, 
    batch_size=32, 
    validation_split=0.2, 
    verbose=0, 
    callbacks=[scale_loss_callback_linear]  
)

y_pred_scaled_linear = best_model_linear.predict(X_test_linear)
y_pred_linear = y_scaler_linear.inverse_transform(y_pred_scaled_linear)
mse_linear = np.mean((y_test_linear.squeeze() - y_pred_linear.squeeze())**2)
r2_linear = 1 - np.sum((y_test_linear.squeeze() - y_pred_linear.squeeze())**2) / np.sum((y_test_linear.squeeze() - np.mean(y_test_linear.squeeze()))**2)

print(f"Test MSE: {mse_linear:.4f}")
print(f"Test R²: {r2_linear:.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Remove irrelevant columns
new_columns_to_drop = ['id', 'name', 'host_id', 'host_name', 'last_review', 
                       'neighbourhood', 'price_bins','log_price','latitude','longitude']
new_data_model = data.drop(columns=new_columns_to_drop)

# One-hot encode categorical variables
new_data_model = pd.get_dummies(new_data_model, 
                                columns=['room_type', 'neighbourhood_group'], 
                                drop_first=False)

# Extract features and target variable
new_X = new_data_model.drop(columns=['price'])
new_y = new_data_model['price']

# Standardize numerical features
scaler = StandardScaler()
new_X_scaled = scaler.fit_transform(new_X)

# Split training and testing sets
new_X_train, new_X_test, new_y_train, new_y_test = train_test_split(
    new_X_scaled, new_y, test_size=0.2, random_state=2
)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import randint

# Define the hyperparameter distributions for Random Forest
param_distributions = {
    'n_estimators': randint(100, 500),    # Random integers between 100 and 500
    'max_depth': randint(10, 20),         # Random integers between 10 and 20
    'min_samples_split': randint(2, 15),  # Random integers between 2 and 15
    'min_samples_leaf': randint(1, 10),   # Random integers between 1 and 10
    'max_features': [1.0, 'sqrt', 'log2'] # Select a fixed subset of features strategy
}

# Initialize the Random Forest Regressor
rf = RandomForestRegressor(random_state=43)

# Define RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_distributions,
    n_iter=25,
    scoring='neg_mean_squared_error',
    cv=5,
    random_state=1,
    n_jobs=-1
)

# Perform hyperparameter search and fit on the new training set
random_search.fit(new_X_train, new_y_train)

print("Best Params:", random_search.best_params_)

# Use the best model for prediction, and name the predictions as new_y_pred
best_rf = random_search.best_estimator_
new_y_pred = best_rf.predict(new_X_test)

# Calculate evaluation metrics
mse = mean_squared_error(new_y_test, new_y_pred)
rmse = mse ** 0.5
r2 = r2_score(new_y_test, new_y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R²: {r2}")
```
